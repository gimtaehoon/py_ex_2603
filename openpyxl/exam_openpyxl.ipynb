```python
from openpyxl import Workbook

w2 = Workbook()

ws = w2.active
ws['A1'] = "Hello"

w2.save('test.xlsx')

ws["c3"] = "bye"

w2.save('test.xlsx')

a='python'

for row in ws['D5':'F8']:
    for cell in row:
        cell.value = a

w2.save('test.xlsx')
```

```python
from openpyxl import load_workbook

wb = load_workbook('test.xlsx')

newws = wb.create_sheet()
newws.title = '마지막_시트'

item1 = {'품목': '바나나','가격':3000,'수량':1,'날짜':'2026년 3월 10일'}
item2 = {'품목': '책','가격':15000,'수량':20,'날짜':'2026년 3월 11일'}
item3 = {'품목': '커피','가격':3500,'수량':10,'날짜':'2026년 3월 12일'}
item4 = {'품목': '신발','가격':50000,'수량':14,'날짜':'2026년 3월 13일'}

items = [item1,item2,item3,item4]

newws['A1']='품목'
newws['B1']='가격'
newws['C1']='수량'
newws['D1']='날짜'

i=2
for item in items:
    newws['A'+str(i)] = item['품목']
    newws['B'+str(i)] = item['가격']
    newws['C'+str(i)] = item['수량']
    newws['D'+str(i)] = item['날짜']
    i+=1

wb.save('test.xlsx')
```

```python
from openpyxl import *
from openpyxl.chart import BarChart, Reference

wb= Workbook()
ws= wb.active

ws.append(['Name','Score'])
ws.append(['홍길동',90])
ws.append(['임꺽정',80])
ws.append(['홍길순',70])
ws.append(['임꺽순',92])

chart = BarChart()

data=Reference(ws, min_col=2, min_row=1, max_row=5)

chart.add_data(data, titles_from_data=True)

chart.title="Student Scores"

ws.add_chart(chart, "E5")

wb.save('chart_exam.xlsx')
```
